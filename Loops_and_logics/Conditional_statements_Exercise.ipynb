```python
# Exercise 1
# categorising animals based on risk status

individuals_in_wild = 5500

if individuals_in_wild < 1000:
    print('Critically Endangered')
elif 1000 <= individuals_in_wild <= 5000:
    print('Endangered')
elif 5001 <= individuals_in_wild <= 20000:
    print('Vulnerable')
else:
    print('safe') 
```

```python
# Exercise 2
# categorise based on rainfall and population

rainfall = 205 # mm
population_density = 448 # per square km

if rainfall < 500 and population_density > 500:
        print('Severe Scarcity')
elif 500 <= rainfall <= 1000 and 200 <= population_density <= 500:
        print('Moderate scarcity')
else:
    print('No Scarcity')
```

```python
# Exercise 3
# categorise based on matrials and size

materials = 'glass'
size = 'medium'

if size == 'large' and materials in ('metal', 'glass'):
    print('Special Handling')
elif size == 'medium':
    print('Standard recycling')
elif size == 'small'and materials in ('metal', 'glass'):
    print('Standard recycling')
elif size == 'small'and materials == 'paper':
    print('Compositing')
elif size == 'small' and materials == 'plastic':
    print('Landfill')
else:
    print('Landfill')
```
