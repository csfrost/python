```python
# Exercise 1
# Print each animal name from the animal list uysing for loop.

animals = ['Great White Shark', 'Blue Whale', 'African Elephant',
           'Bald Eagle', 'Orangutan', 'Tiger', 'Panda', 'Koala']

for i in animals:
    print(i)
```

```python
# Exercise 2
# Use the while loop to print numbers from 1 to 4

numbers = 0

while numbers < 5:
    print(numbers)
    numbers += 1
```

```python
# Exercise 3
# Stop the loop using BREAK once you find Orangutan in the animals list.

for i in animals:
    print(i)
    if i == "Orangutan":
        print(f'{i} has been found')
        break # Stops the process from going further
```

```python
# Exercise 4
# Print all animals except 'Bald Eagle' using the for loop and continue statement

for i in animals:
    if i == 'Bald Eagle':
        continue # Skips Bald Eagle if found and continues the process
    print(i)
```

```python
# Exercise 5
# Create a loop that does nothing when 'Panda' is encountered.

for i in animals:
    if i == 'Panda':
        pass # does nothing, mostly act as a placeholder
    else:
        print(i)
```
