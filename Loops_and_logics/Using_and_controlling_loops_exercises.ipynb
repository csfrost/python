### Exercise 2
##### Use a while loop to simulate a deforestation monitoring system that alerts when the forested area falls below a critical threshold.

```python
initial_forested_area = 1000 # in square kilometres
deforestation_rate = 20 # square kilometres per year

# Task: how many years will it take for the forested area to fall below 500 square kilometres
years = 0
while initial_forested_area > 500:
    initial_forested_area -= deforestation_rate
    years +=1

print(f"Years until critical deforestation level: {years} years")
```

### Exercise 3
##### Write a for loop to analyse a list of animals, printing out wether they are endangered or not.

```python
animals = ['Tiger', 'Blue Whale', 'African Elephant', 'Koala', 'Panda']
endangered_animals = ['Tiger', 'Blue Whale', 'African Elephant']

# Task: Use a for loop to print each animal's name and its endangered status
for animal in animals:
    if animal in endangered_animals:
        print(f"Animal name: {animal}\n\tStatus: Endangered animal")
    else:
        print(f"Animal name: {animal}\n\tStatus: Not an endangered animal")
```

### Exercise 4
##### Use loops to process and display informations about ocean pollution levels in different regions

```python
ocean_pollution_data = {
    'Pacific Ocean': [3, 5, 2],
    'Atlantic Ocean': [7, 2, 4],
    'Indian Ocean': [5, 1, 3]
}

# Task: use loops to calculate the average pollution level for each ocean
for oceans, pollutions in ocean_pollution_data.items():
    sum_pollution_level = sum(pollutions)
    average_pollution_level = sum_pollution_level/len(pollutions)

    print(f"{oceans} pollution level = {sum_pollution_level}\n\t average pollution level = {average_pollution_level:.2f}")
```

### Exercise 6: Basic list comprehension
##### Fill the comprehension line to create a list of squares of numbers from 1 to 10.

```python
numbers = range(1, 11)
squares = [x**2 for x in numbers]
print(squares)
```

### Exercise 7: List comprehension with a conditional
#### Use list comprehension to create a list of only even numbers from 1 to 20

```python
numbers = range(1,21)
even_numbers = [x for x in numbers if x%2 == 0]
print(even_numbers)
```

### Exercise 8: dictionary comprehension
##### Create a dictionary comprehension that categorises each marine animal in a list as "Aquatic" using dictionary comprehension

```python
marine_animals = ['Dolphin', 'Shark', 'turtle']
animal_categories = {animal: 'Aquatic' for animal in marine_animals}
print(animal_categories)
```

### Exercise 9: Data mapping with dictionary comprehension

```python
species = ['Dolphin', 'Whales', 'Sea Turtle', 'Seals']
status = ['Endangered', 'Vulnerable', 'Endangered', 'Least Concern']

number_of_species = range(len(species))
endangered_species = {species[x]:status[x] for x in number_of_species}
print(endangered_species)
```
