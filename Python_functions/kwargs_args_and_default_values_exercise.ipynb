### Exercise 1
Create a Python function named `calculate_forest_area` that accepts two arguments: `trees` (the number of trees) and `area_per_tree` (the space each tree occupies, defaulting to 10 square metres). The function should compute the total forested area based on these inputs. It should also include an optional `deforestation_factor` parameter, which defaults to **0.2** and represents the proportion of forest lost to deforestation. The function should return the **remaining forest area** after subtracting the deforested portion.

```python
def calculate_forest_area(trees, area_per_tree =10, deforestation_factor=0.2):
    """
    Calculates and return the net forest area 
    after considering deforestation

    Parameters:
    -------
    - trees: integer, Number of trees in the forest 
    - area_per_tree: float, area of forest covered per tree 
                    (default value is 10 square metres)
    - deforestation_factor: float(optional), default deforestation value

    Returns
    -------
    net_area: float(optional), represents the overall area of the forest 
                       covered by the trees after considering deforestation
    """
    # Calculate the total area covered by the trees in the forest
    total_area_covered = trees * area_per_tree 

    # calculate the net area after deforestation
    net_area = total_area_covered * (1-deforestation_factor)
    
    return net_area
```

Using the above function, calculate the area of a forest with **50 trees**, assuming each tree covers **5 square metres**, with a **deforestation factor of 0.3**. Print the result.

```python
# Using the function calculate_forest_area to calculate the area of the given forest
calculate_forest_area(50, 5, 0.3)
```

### Exercise 2

Write a function called `tree_planting_campaign` that monitors the progress of tree planting initiatives. The function should accept two parameters: `initial_trees` (the starting tree count) and `trees_per_campaign` (how many trees are planted during each campaign). The function must allow users to **input tree counts for any number of campaigns, depending on how many actually took place**. It should return the cumulative total of trees once all campaigns are complete.

```python
def tree_planting_campaign(initial_trees, *trees_per_campaign):
    """
    Tracks the progress of tree planting campaign

    Parameter
    -------
    - initial_trees: int, Initial number of trees before the campaign
    - tree_per_campaign: int, number of trees planted per campaign

    Return
    -------
    - total_number_of_trees: Final number of trees after the campaign
    """

    for campaign in trees_per_campaign:
        initial_trees += campaign
        total_number_of_trees = initial_trees
    return total_number_of_trees
```

Using the above function, track and **print** the progress of a few tree planting campaigns. The forest started with **100** trees. In subsequent campaigns, the first campaign produced **20** additional trees, the second **30**, and the third **50** trees.

```python
# Using the function tree_plantain_campaign to track and print the progress of the campign 
tree_planting_campaign(50, 10, 50, 40, 20, 80)
```

### Exercise 3

Write a function called `deforestation_impact_report` that produces a summary of deforestation effects. The function should **accept a flexible number of impact categories along with their associated values**, representing various consequences of deforestation like "loss_of_biodiversity," "carbon_emission," and so on. The function should **display each impact category alongside its corresponding value in a clearly labelled format**.

```python
def deforestation_impact_report(**impact_of_deforestation):
    """
    Generates a report on the impact of deforestation
    
    """
    for aspects, impacts in impact_of_deforestation.items():
        print(f"{aspects}: {impacts}")
```

Using the above function, create a report on the impact of deforestation, noting a **30%** loss of biodiversity, a carbon emission of **500** units, and soil erosion of **20** units.

```python
deforestation_impact_report(
    loss_of_biodiversity = 30, 
    carbon_emission_value = 500, 
    soil_erosion = 20
)
```
